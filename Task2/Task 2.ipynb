```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
# This approach with networkx library is proposed by me (not vanila), as it is gentle and clear understandable
class Graph:
    def __init__(self):
        self.edges = [] # the set of edges
        self.nodes = [] # the set of nodes
        
    def add(self, a, b): # MULTITASK function to add edge if a!=b, and to add node if a==b;  А, Б mini tasks
        if a!=b:
            self.edges.append([a, b])
            if a not in self.nodes: self.nodes.append(a)
            if b not in self.nodes: self.nodes.append(b)
        else: 
            self.nodes.append(a)
            
    def visualize(self): # plt.show() - display the graph with matplotlib as required; Г mini task
        graph = nx.Graph()
        graph.add_edges_from(self.edges)
        graph.add_nodes_from(self.nodes)
        nx.draw_networkx(graph)
        plt.show()
```

```python
G = Graph()
G.add("A", "B")
G.add("C", "D")
G.add("C", "B")
G.add("E", "B")
G.add("E", "C")
G.add("D", "E")
G.add("B", "K")
G.visualize()
```

```python
class random_Graph: # В mini task
    def __init__(self):
        self.edges = [] # the set of edges
        self.nodes = [] # the set of nodes
        
    def gen(self, n=random.randint(10, 20), X='ABCDEFJ'): # function to add edge if a!=b, and to add node if a==b 
        for i in range(n):
            a,b=random.choice(X),random.choice(X)
            if a!=b:
                self.edges.append([a, b])
                if a not in self.nodes: self.nodes.append(a)
                if b not in self.nodes: self.nodes.append(b)
            else: 
                self.nodes.append(a)
            
    def visualize(self): # plt.show() - display the graph with matplotlib as required
        graph = nx.Graph()
        graph.add_edges_from(self.edges)
        graph.add_nodes_from(self.nodes)
        nx.draw_networkx(graph)
        plt.show()
```

```python
# lets create several RANDOM graphs
```

```python
G=random_Graph() # initialize graphs
G.gen() # add nodes and ages
```

```python
G.visualize() # visualise 
```

“Вы выбираете из группы из N друзей, с кем пойти на пикник. Однако друзья, связанные
ребром на графе, находятся в очень недружественных отношениях друг с другом,
поэтому, если они оба отправятся на пикник, он будет испорчен. 

Чтобы на пикнике было как можно больше друзей, кого следует пригласить?. 

Протестируйте алгоритм на случайных графах (5 - 10 штук) различного размера, отрисуйте случайные графы и
выведите финальный список на экран или в фаил.”

The intuition for  solution is lies in the next topic:
   We have to invite friends with less connections with other people firstly (degree of a vertex is --> min), 
   and than only invite people with bigger and bigger degree of a the vertex. 
    
   So, degree of each vertex should be precomputed. Such approach allows to invite most people, which are not familar with each other
   
   (degree of a vertex - number of edges for the node).
